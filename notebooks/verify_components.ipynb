## Check Sentence Transformer

```python
from sentence_transformers import SentenceTransformer
import time

print("Loading embedding model...")
start = time.time()

# This downloads the model (~80MB) on first run
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

end = time.time()
print(f"âœ… Model loaded in {end-start:.2f} seconds")

# Test embedding
test_vec = model.encode("Test sentence")
print(f"âœ… Generated embedding: dimension={len(test_vec)}")
```
